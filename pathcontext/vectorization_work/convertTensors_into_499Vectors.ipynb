```python
import numpy as np
import pandas as pd

sol_data = pd.read_csv('/home/supriya/pathcontext_CDS_v2x/v2/CDS_100fv.csv',dtype={'label':object} )
id=sol_data["ID"]
label=sol_data["label"]
length=len(sol_data["Vector_form"])
print(sol_data)
sol_data.head()
```

```python
import re
z=''
for j in range(201):
    if j==0:
        continue
    elif j!=200:
        z=z+'v'+str(j)+','
    else:
        z=z+'v'+str(j)
#print(z)
savefile = open('/home/supriya/pathcontext_CDS_v2x/v2/CDS_200fv.txt', 'a')
savefile.write(str('ID') +","+str(z)+ ","+'label'+'\n')
savefile.close()
    
for i in range(length):
    
    
    x = sol_data["Vector_form"][i]
    m = re.search('\[([^]]+)\]', x)
    if m:
        x = m.group(1)
    
#print(x)
    y=re.sub('[+\s]','', x)

    savefile = open('/home/supriya/pathcontext_CDS_v2x/v2/CDS_200fv.txt', 'a')
    savefile.write(str(id[i]) +","+str(y)+ ","+ str(label[i])+'\n')
    savefile.close()
```
