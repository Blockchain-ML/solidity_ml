```python
import numpy as np
import pandas as pd

sol_data = pd.read_csv('/home/supriya/pathcontext/batch4/py_example/TrainB7_800.csv',dtype={'Error_Label':object} )
id=sol_data["ID"]
label=sol_data["Error_Label"]
length=len(sol_data["Vector_form"])
print(sol_data)
sol_data.head()
```

```python
import re
z=''
for j in range(500):
    if j==0:
        continue
    elif j!=499:
        z=z+'v'+str(j)+','
    else:
        z=z+'v'+str(j)
#print(z)
savefile = open('/home/supriya/split_cols_v2/ValBatch4.txt', 'a')
savefile.write(str('ID') +","+str(z)+ ","+'Error_Label'+'\n')
savefile.close()
    
for i in range(length):
    
    #x = sol_data["Vector_form"][1][8:-26]
    x = sol_data["Vector_form"][i]
    m = re.search('\[([^]]+)\]', x)
    if m:
        x = m.group(1)
    
#print(x)
    y=re.sub('[+\s]','', x)

    savefile = open('/home/supriya/split_cols_v2/ValBatch4.txt', 'a')
    savefile.write(str(id[i]) +","+str(y)+ ","+ str(label[i])+'\n')
    savefile.close()
```
