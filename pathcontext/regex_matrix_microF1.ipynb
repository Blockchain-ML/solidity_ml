```python
import re
import sys
import glob
import os.path

list_of_files = glob.glob(r"/home/supriya/pathcontext/MixModel/UNB_UNB/*.txt") 
for file_name in list_of_files:
    filename=os.path.splitext(os.path.basename(file_name))[0]
#     print(filename)
    f = open(file_name,"r")
#     f = open('./xgb_smt_svm.txt')

    content = f.read()

    matrices = re.findall(r"\[\[[\s0-9\s]+\]\n\s\[[\s0-9\s]+\]\]", content) 

#     print(matrices)

    # RegEx work Complete

    # Generate Matrices
    # tn, fp, fn, tp
    sum_tp = 0;  
    sum_fn =0 
    sum_fp = 0
    sum_tp_fp=0
    sum_tp_fn=0
    TP=[]
    FP=[]
    FN=[]
    for matrix in matrices:
        matrix = matrix.replace('[', '')
        matrix = matrix.replace(']', '')
        matrix = matrix.replace('\n', '')
        mat = re.sub(r'\s+', ',', matrix.strip())

        mat = mat.split(',')
    #     print(mat)
        fp = mat[1]
        fn = mat[2]
        tp = mat[3]
    #     print(fp,fn,tp)
        TP.append(tp)
        FP.append(fp)
        FN.append(fn)
#     print(TP)  
#     print(FP)
#     print(FN)
    # print(type(TP))
    for i in range(0, len(TP)):    
        sum_tp = sum_tp + int(TP[i])
        sum_fp = sum_fp + int(FP[i])
        sum_fn = sum_fn + int(FN[i])  


    sum_tp_fp =sum_tp + sum_fp
    sum_tp_fn=sum_tp + sum_fn

    micro_precision = sum_tp/sum_tp_fp
    micro_recall = sum_tp/sum_tp_fn
    micro_avg=(micro_precision+micro_recall)/2
#     print(micro_precision,micro_recall)
#     print(micro_avg)
    print(str(filename)+"\n"+"Micro_precision:  "+str(micro_precision) +"\n"+"Micro_recall:  "+str(micro_recall) +"\n"+"Micro_avg:  "+str(micro_avg) +"\n")
        # Save Results as txt
    filename1 = f'/home/supriya/pathcontext/MixModel/Micro_AVG_unb_unb.txt'
    with open(filename1, 'a') as fh:
        fh.write(str(filename)+"\n"+"Micro_precision:  "+str(micro_precision) +"\n"+"Micro_recall:  "+str(micro_recall) +"\n"+"Micro_avg:  "+str(micro_avg) +"\n"+"\n")


```
