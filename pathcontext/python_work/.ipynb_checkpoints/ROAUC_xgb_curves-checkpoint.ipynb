```python
import numpy as np
import pandas as pd
import sklearn
from sklearn.metrics import classification_report,accuracy_score, f1_score, recall_score, precision_score, confusion_matrix
from sklearn import tree
from sklearn import preprocessing
from matplotlib import pyplot
from sklearn.model_selection import cross_val_score 
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve, roc_auc_score
import pickle
from sklearn import metrics
# open a file, where you stored the pickled data
loaded_model = pickle.load(open('/home/supriya/pickle/l7_model_9.pickle','rb'))
X_smv = pickle.load(open('/home/supriya/pickle/l7_X_9.pickle','rb'))
y_smv = pickle.load(open('/home/supriya/pickle/l7_Y_9.pickle','rb'))
y_predict = loaded_model.predict(X_smv)
pd.crosstab(y_smv,y_predict)
# print(loaded_model)
y_predict = loaded_model.predict(X_smv)
pd.crosstab(y_smv,y_predict)
yhat = loaded_model.predict_proba(X_smv)
# keep probabilities for the positive outcome only
yhat = yhat[:, 1]

# print AUROC is the area under the ROC curve
l7_auc = roc_auc_score(y_smv, y_predict)
print('Label 5 : AUROC = %.3f' % (l7_auc))

# calculate roc curves
fpr, tpr, thresholds = metrics.roc_curve(y_smv, yhat)

# # plot the roc curve for the model
# pyplot.plot([0,1], [0,1], linestyle='--')
# pyplot.plot(fpr, tpr,  marker='.', label=' Label 5 XGBoost (AUROC = %0.3f)' % l7_auc)

# # Title
# pyplot.title('ROC Plot')
# # Axis labels
# pyplot.xlabel('False Positive Rate')
# pyplot.ylabel('True Positive Rate')
# # Show legend
# pyplot.legend() # 
# # Show plot
# pyplot.show()
```
