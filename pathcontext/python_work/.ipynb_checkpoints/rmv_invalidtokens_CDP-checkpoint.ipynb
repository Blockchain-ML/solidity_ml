```python
import numpy as np
import pandas as pd
import subprocess
import json
import os
import re 
data = pd.read_csv(r'/home/supriya/CDP_MIT/test/CDP2.csv')
# data = pd.read_csv(r'/home/supriya/CDP_MIT/path_exp/CDP/CDPtest.csv')
df = pd.DataFrame(data)
df_new = df.dropna()
print(df.shape)
print(df_new.shape)
ts2=[]
te2=[]
t1=df_new["tokenstart"]
t2=df_new["tokenend"]
for ts1 in t1:
    pattern = "([a-zA-Z$_][a-zA-Z0-9$_]*)"
    test_string = ts1
    result = re.search(pattern, str(test_string))
#     print(result ,",",ts1)
    if result is not None:
        ts2.append(ts1)
# print(ts2) 
for te1 in t2:
    pattern = "([a-zA-Z$_][a-zA-Z0-9$_]*)"
    test_string = te1
    result = re.search(pattern, str(test_string))
#     print(result ,",",ts1)
    if result is not None:
        te2.append(te1)
# print(te2)        
new_df=df_new[df_new.tokenstart.isin(ts2)]
new_end=new_df[new_df.tokenend.isin(te2)]
# new_df=df[~df.tokenstart.str.contains(ts2)]
# print(new_end)
df_data_list=['None', 'solidity','pragma','function',"to",'<EOF>','returns','mapping','from','contract','transfer','external','pure'
             ,'view','public','after','case', 'default','final','in','inline','let','match','null','of','relocatable','static'
  ,'switch','typeof','anonymous','break','constant','immutable','continue','leave','external','indexed','internal','payable',
              'private','public','virtual' ,'pure' ,'type','view','constructor','fallback' ,'receive','wei' , 'szabo' ,'finney','ether'
,'seconds','minutes','hours','days','weeks','years',"modifier","struct" ,'override','event','enum','calldata','storage','emit',
'throw','continue','break','new','after','delete','from','assembly','return','count','if','for','while','let','true','false',
             'do','continue','try','else','using','memory','override','abstract','contract','interface','library','import','is',
             'as','value',]
new_end1=new_end[~new_end.tokenstart.isin(df_data_list)]
new_end2=new_end1[~new_end1.tokenend.isin(df_data_list)]
# print(new_end2)
num2=[]
num3=[]
for num1 in t1:
    regex_num="0[xX][0-9a-fA-F]+"
    test_string = num1
    result = re.search(regex_num, str(test_string))
#     print(result ,",",ts1)
    if result is not None:
        num2.append(num1)
for hex1 in t2:
    regex_num="0[xX][0-9a-fA-F]+"
    test_string = hex1
    result = re.search(regex_num, str(test_string))
#     print(result ,",",ts1)
    if result is not None:
        num3.append(hex1)        
new_end11=new_end2[~new_end2.tokenstart.isin(num2)]
new_end21=new_end11[~new_end11.tokenend.isin(num3)]
print(new_end11.shape)
print(new_end21.shape)
# print(df1)
# print(num3)
print(new_end21)
```
