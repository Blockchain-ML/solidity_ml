```python
from selenium import webdriver
import pandas as pd
import time
import re
import os
import os.path

if os.path.isfile('Error_label.text'):
    os.remove('Error_label.text')
driver = webdriver.Chrome(executable_path=r"C:\Users\hp\PycharmProjects\Selenium_labelling\drivers\chromedriver.exe")

data = pd.read_csv(r"C:\Users\hp\PycharmProjects\Selenium_labelling\Test\out_new.csv")
depth = len(data['solidity']) #this finds how deep the columns are
id=len(data['_id'])
ids=data['_id']
#print(ids)
print(depth);
sol=(data['solidity']) ;
print(sol);

driver.implicitly_wait(10)
driver.get('https://tool.smartdec.net/newscan?type=codeblock')
#driver.maximize_window()
errors = ["Unsafe array's length manipulation", 'Costly loop', 'Timestamp dependence',
          'Use of call function with no data', 'Blockhash function misuse', 'Using tx.origin for authorization',
          'Non-initialized return value', 'Hardcoded address', 'Checking for strict balance equality',
          'Token API violation', 'DoS by external contract', 'Locked money', 'Integer division', 'Malicious libraries',
          'Reentrancy', 'Strict comparison with block.timestamp or now', 'Overflow and Underflow in Solidity',
          'Redundant fallback function', 'Revert inside the if-operator',
          'Replace multiple return values with a struct', 'Transfer forwards all gas', 'Extra gas consumption',
          'Overpowered role', 'Send instead of transfer', 'ETH transfer inside the loop']


def generate_error_label(contract_errors):
    error_label = ['0'] * len(errors)
    for error in contract_errors:
        found= error in errors
        if(found):
            pos = errors.index(error)
            error_label[pos] = '1'
    return ''.join(error_label)


for i in range(0, depth):
    if i==3 or i==15 or i==38 or i==42:
        continue
    os.system('echo "{}"|xlip -sectional clipboard'.format(sol[i]))

    elem = driver.find_element_by_id("codeblock")
    elem.send_keys((sol[i]));

    time.sleep(5)
    driver.find_element_by_class_name('btn-submit').click()

    time.sleep(10);
    driver.get(driver.current_url)
    time.sleep(10)
    html_source = driver.page_source
    time.sleep(10)
    ans = re.findall('<div _ngcontent-c1=\"\"\>([^<]*)<\/div>', html_source)
    print(ans)
    res = [sub for sub in ans if sub in errors]
    val = str(res)
    print('Matched_error:'+val)
    print(str(i)+','+' '+str(generate_error_label(ans)))

    time.sleep(5)
    savefile = open('Error_label.text', 'a')
    savefile.write(str(i)+','+' '+str(generate_error_label(ans))+'\n')
    savefile.close()

    time.sleep(5)

    driver.get('https://tool.smartdec.net/newscan?type=codeblock');

    time.sleep(5);

driver.close()





```
