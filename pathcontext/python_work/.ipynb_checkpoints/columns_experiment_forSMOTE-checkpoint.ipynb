```python
#multi cols of labels
#Create independent and Dependent Features
#data=split_label(data)
def split_label(data_frame):
    # Change 'label' to whatever naming used in the original DataFrame.
    temp_df = data_frame['label'].str.split("", n = -1, expand = True)
    for i in range(1, 19):
        data_frame[f'l{i}'] = temp_df[i]
    return data_frame

newdata=split_label(data)
le = preprocessing.LabelEncoder()
balance_data = newdata.apply(le.fit_transform)
#Y=balance_data
X = balance_data.values[:, :4]

Y = balance_data.values[:,4:]
# Print the shapes of X & Y
print(X)
print(X.shape)
print(Y)
print(Y.shape)
```

```python
#multi class label binarizer
def split_label_binarizer(data_frame):
    # Change 'label' to whatever naming used in the original DataFrame.
    temp_df = data_frame['label'].str.split("", n = -1, expand = True)
    
    #import numpy as np
    #from sklearn import preprocessing
    lb = preprocessing.LabelBinarizer()
    template = ['0', '1']
    template = [int(i) for i in template]  
    lb.fit_transform(np.array([template]))
    # print(lb.classes_)
    
    cl = []
    for index in data_frame.index:
        temp = data_frame['label'][index]
        temp = [int(i) for i in temp] 
        # print(temp)
        # temp = temp[1:19]
        mat = lb.transform(temp)
        cl.append(mat)
    data_frame['lc'] = cl
    return data_frame
newdata=split_label_binarizer(data)
#print(newdata)

X = newdata.values[:, 0:4]

Y = newdata.values[:,4:]
# Print the shapes of X & Y
#print(X)
#print(X.shape)
print(Y)
print(Y.shape)
```
