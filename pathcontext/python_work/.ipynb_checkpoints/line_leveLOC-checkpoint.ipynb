```python
import sys
import glob
import os.path

filename = r"/home/supriya/pathcontext/temp/55567.txt"

errors = ['SOLIDITY_ARRAY_LENGTH_MANIPULATION', 'SOLIDITY_GAS_LIMIT_IN_LOOPS','SOLIDITY_CALL_WITHOUT_DATA',
          'SOLIDITY_INCORRECT_BLOCKHASH', 'SOLIDITY_TX_ORIGIN','SOLIDITY_ADDRESS_HARDCODED', 'SOLIDITY_BALANCE_EQUALITY',
          'SOLIDITY_ERC20_FUNCTIONS_ALWAYS_RETURN_FALSE','SOLIDITY_LOCKED_MONEY', 'SOLIDITY_EXACT_TIME',
          'SOLIDITY_REDUNDANT_FALLBACK_REJECT', 'SOLIDITY_REVERT_REQUIRE','SOLIDITY_SHOULD_RETURN_STRUCT',
          'SOLIDITY_DEPRECATED_CONSTRUCTIONS', 'SOLIDITY_EXTRA_GAS_IN_LOOPS',
          'SOLIDITY_OVERPOWERED_ROLE', 'SOLIDITY_SEND', 'SOLIDITY_TRANSFER_IN_LOOP']
LOC = 185

f= open(filename, 'r')
#print(f.read())
error_list=[]
count = 0
ln=0
ln_error_dict={}


def generate_error_label(error_list):
    #print('hi')
    error_label = ['0'] * len(errors)
    for error in error_list:
#             print(error)
        found= error in errors

        if(found):
            #print(found)
            pos = errors.index(error)
            error_label[pos] = '1'
    return ''.join(error_label)

for line in f:
        if line.startswith(('ruleId')):
            lst=line
            s1=lst.rstrip().split(':')
            v1=s1[:2]
            v2=v1[1]
            v2 = v2.strip()
        elif line.startswith('line'):
            lst1=line
            s2=lst1.rstrip().split(':')
            v4=s2[:2]
            #print(v1)
            v3=v4[1]
            v3 = v3.strip()
            if(ln_error_dict.get(v3) == None):
                ln_error_dict[v3]=v2
            else:
                temp=ln_error_dict.get(v3)
                ln_error_dict[v3]=temp+','+v2
                
for key, value in ln_error_dict.items():
    error_list=value.split(',')
    error_label=generate_error_label(error_list)
    savefile = open(r'/home/supriya/pathcontext/temp/lineby55567.txt', 'a')
    savefile.write(str(key)+","+str(error_label)+","+value+'\n')
    savefile.close()

count=1
default_err_label='000000000000000000'

while count <= LOC:
    if(ln_error_dict.get(str(count)) == None):
        savefile = open(r'/home/supriya/pathcontext/temp/lineby55567.txt', 'a')
        savefile.write(str(count)+","+default_err_label+'\n')
        savefile.close()
    count=count+1
            
f.close()

```
