```python
import re
import sys
import glob
import os.path

# list_of_files = glob.glob(r"/home/supriya/pathcontext/cw_soliaudit/smotetomek-svmsmote/*.txt") 
list_of_files = glob.glob(r"/home/supriya/pathcontext/MixModel/smt_svm/*.txt") 
for file_name in list_of_files:
    filename=os.path.splitext(os.path.basename(file_name))[0]
#     print(filename)
    f = open(file_name,"r")
#     f = open('./xgb_smt_svm.txt')

    content = f.read()

    matrices = re.findall(r"\[\[[\s0-9\s]+\]\n\s\[[\s0-9\s]+\]\]", content) 
            # Save Results as txt
    filename1 = f'/home/supriya/pathcontext/MixModel/TPRs_TNRs1.txt'
    with open(filename1, 'a') as fh:
        fh.write("\n"+str(filename)+"\n")

#     print(matrices)

    # RegEx work Complete

    # Generate Matrices
    # tn, fp, fn, tp
    sum_tpr = 0 
    sum_fnr =0 
    sum_fpr = 0
    sum_tnr=0

    TPr=[]
    FPr=[]
    FNr=[]
    TNr=[]
    count=0
    for matrix in matrices:
        count = count+1
        matrix = matrix.replace('[', '')
        matrix = matrix.replace(']', '')
        matrix = matrix.replace('\n', '')
        mat = re.sub(r'\s+', ',', matrix.strip())

        mat = mat.split(',')
    #     print(mat)
        tn = mat[0]
        fp = mat[1]
        fn = mat[2]
        tp = mat[3]
#         print(fp,fn,tp)
        sum_tp_fn1 = int(tp)+int(fn)
        sum_tn_fp1= int(tn)+int(fp)
        sum_fp_tn1 = int(fp) + int(tn)
#         print(type(sum_tp_fn))

#  true positive rate
#         TPR = TP/(TP+FN)
        # Specificity or true negative rate
#         TNR = TN/(TN+FP) 
        # False negative rate
        # FNR = FN/(TP+FN)
        # Fall out or false positive rate
#         FPR = FP/(FP+TN)
        
        TNR = int(tn) /sum_tn_fp1
        TPR = int(tp) / sum_tp_fn1
        FNR = int(fn) / sum_tp_fn1
        FPR = int(fp) / sum_fp_tn1
        

        print("TPR:"+str(count )+" "+str(TPR)+" "+"TNR:"+str(count )+"  "+str(TNR)+" "+"FPR:"+str(count )
              +"  "+str(FPR)+" "+"FNR:"+str(count )+"   "+str(FNR))
        
        filename1 = f'/home/supriya/pathcontext/MixModel/TPRs_TNRs1.txt'
        with open(filename1, 'a') as fh:
            fh.write("TPR:"+str(count )+" "+str(TPR)+" "+"TNR:"+str(count )+"  "+str(TNR)+" "+"FPR:"+str(count )
                     +"  "+str(FPR)+" "+"FNR:"+str(count )+"   "+str(FNR)+"\n")
   
        TPr.append(TPR)
        FPr.append(FPR)
        FNr.append(FNR)
        TNr.append(TNR)

    for i in range(0, len(TPr)): 
#         print(TPr[i])
        sum_tpr = sum_tpr + float(TPr[i])
        sum_fpr = sum_fpr + float(FPr[i])
        sum_fnr = sum_fnr + float(FNr[i])  
        sum_tnr = sum_tnr + float(TNr[i])
        
#     print(sum_tpr)
    avg_TPR = sum_tpr/len(TPr)
    avg_TNR = sum_tnr/len(TNr)
    avg_FPR = sum_fpr/len(FPr)
    avg_FNR = sum_fnr/len(FNr)
#     print(avg_TPR)
    print("\n"+"\n"+str(filename)+"\n"+"AVG_TPR: "+str(avg_TPR)+" "+"AVG_TNR: "+str(avg_TNR)+" "+"AVG_FPR: "+str(avg_FPR)+" "+"AVG_FNR: "+str(avg_FNR)+"\n")


# #         # Save Results as txt
    filename1 = f'/home/supriya/pathcontext/MixModel/TPRs_TNRs1.txt'
    with open(filename1, 'a') as fh:
        fh.write("\n"+"AVG_TPR: "+str(avg_TPR)+" "+"AVG_TNR: "+str(avg_TNR)+" "+"AVG_FPR: "+str(avg_FPR)+" "+"AVG_FNR: "+str(avg_FNR)+"\n"+"\n")


```
