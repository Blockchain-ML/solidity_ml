```python
class Employee:
    num_of_emps = 0
    raise_amount = 1.04   #---CLASS VARIABLE---------->
    def __init__(self,first,last,pay):
        self.first = first
        self.last = last
        self.email = first+'.'+last+'@iitp.ac.in'
        self.pay = pay
        Employee.num_of_emps+=1
        
    def fullname(self):
        return'{} {}'.format(self.first, self.last)  
    
    def apply_raise(self):
# WHEN WE ACCESS THESE CLASS VARABLE RAISE_AMOUNT WE NEED EITHER ACCESS THEM THROUGH THE CLASS itself or INSTANCE OF THE CLASS        
# WRONG----->self.pay = int(self.pay*raise_amount)
      self.pay = int(self.pay * self.raise_amount)
    
class Developer(Employee):
    raise_amount = 1.10
    
    def __init__(self,first,last,pay, prog_lang):
        super().__init__(first,last,pay)  #but super is maintanable
#         Employee.__init__(self,first,last,pay) #----both ways are same of calling the parent class using __init__ method will work
        self.prog_lang = prog_lang
    pass

class Manager(Employee):
    def __init__(self,first,last,pay, employees=None): 
        super().__init__(first,last,pay)
        if employees is None :
            self.employees=[]
        else:
            self.employees=employees
    
    def add_emp(self,emp):
        if emp not in self.employees:
            self.employees.append(emp)
                
    def remove_emp(self,emp):
        if emp in self.employees:
            self.employees.remove(emp)
    
    def print_emps(self):
        for emp in self.employees:
            print("----->", emp.fullname())
            
        
            

# dev_1 = Employee('manish','prasad',100000)   #AFTER applying raise_amt 1.04 of employee class
dev_1 = Developer('manish','prasad',100000,'javascript')  #AFTER applying raise_amt 1.10 of developer class
dev_2 = Developer('promita','shakya',50000,'python')
dev_3 =Developer('tayoki','shakya',50000,'C++')
      
        
# emp_1 = Employee('supriya','shakya',500000)
# emp_2 = Employee('khushi','shakya',600000) 

mgr_1=Manager("Riya",'Shakya',130000,[dev_1])
mgr_1.add_emp(dev_2)
mgr_1.remove_emp(dev_1)

#isinstance() will tell us if an object is an instance of a class(True or False)
print(isinstance(mgr_1,Manager ))
print(isinstance(mgr_1,Developer ))
print(isinstance(mgr_1,Employee ))

#issubclass() function will tell us is a class is a  subclass of another or not
print(issubclass(Developer,Manager ))
print(issubclass(Developer,Developer ))
print(issubclass(Developer,Employee ))


# print(dev_1.email)
print(dev_1.prog_lang)
# print(dev_2.email)
# print(help(Developer))

print(dev_1.pay)
dev_1.apply_raise()   #AFTER applying raise_amt on specific instance 
dev_2.apply_raise()
print(dev_1.pay)
print(dev_2.pay)

print(mgr_1.email)
print(mgr_1.print_emps())
```
