```python
from queue import Queue

adj_list = {
 "A":["B" , "D"],
 "B":["A", "C"],
 "C":["B"],
 "D":["A", "E","F"],
 "E":["D","F","G"],
 "F":["D","E","H"],
 "G":["E","H"],
 "H":["G","F"]
}


#Breath first search  CODE
visited = {}
level = {}
parent = {}
bfs_traversal_output = []
queue = Queue()
for node in adj_list.keys():
    visited[node] = False
    parent[node] = None
    level[node] = -1
    
print(visited)  
print(parent)
print(level)
    
source = "A"
visited[source] = True
level[source] = 0
queue.put(source)
while not queue.empty():
    u = queue.get()  #pop the visited element of queue
    print("pop the first element of queue is " + u)
    bfs_traversal_output.append(u)
#     print(bfs_traversal_output)
    
    #explore adjacent vertices of u
    for v in adj_list[u]:
        if not visited[v]:
            visited[v] = True
            parent[v] = u
            level[v] = level[u] +1
            print(f"Level is {v} " + str(level[v]))
            queue.put(v)
print("bfs_traversal_output is " + str(bfs_traversal_output)     )       
```
